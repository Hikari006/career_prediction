```python
x = df.drop(columns=['Career'])
y = df['Career']
```

```python
x.boxplot(rot=40)
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()
cols = x.columns
X = ms.fit_transform(x)
X = pd.DataFrame(X, columns=[cols])
X.head()
```

kmean

```python
from sklearn.cluster import KMeans

clusters = 5
km = KMeans(n_clusters=clusters, n_init = 10)
km.fit(X)

pd.Series(km.labels_).unique()
```

```python
df['Groups_2'] = km.labels_
df.sort_values('Groups_2', inplace=True)
df.head()
```

Clustering Groups by KMeans

```python
for i in pd.Series(km.labels_).unique():
    careers =df[df['Groups_2'] == i]['Career'].values
    print(f'\n- - - Group {i} with size {len(careers)}:\n')
    print(careers)
```

```python
x_big5=df.drop(["Numerical Aptitude","Spatial Aptitude","Perceptual Aptitude","Abstract Reasoning","Verbal Reasoning","Career","Groups_2"],axis="columns")
y_big5=df.Groups_2
```

```python
x_big5
```

```python
y_big5
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train_big5,y_test_big5=train_test_split(x_big5,y_big5,test_size=0.2,random_state=42)
```

```python
len(X_train)
```

knn

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train_big5)
```

```python
knn.score(X_test,y_test_big5)
```

```python
knn.score(X_train,y_train_big5)
```

```python
y_pred_big5=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acuraccy=accuracy_score(y_test_big5,y_pred_big5)
print(acuraccy)
```

```python
from sklearn.preprocessing import StandardScaler
new_data=[7.5,5.6,0.6,5.4,7.7]


pred_career=knn.predict([new_data])
print(f'predicted career :{pred_career}')
```

```python
from sklearn.metrics import confusion_matrix

y_pred_big5=knn.predict(X_test)
cm=confusion_matrix(y_test_big5,y_pred_big5)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test_big5, y_pred_big5))
```
